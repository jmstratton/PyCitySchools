# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 
---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data = pd.DataFrame(school_data)
student_data = pd.DataFrame(student_data)

# Preview Data
# school_data.head()
student_data.head()

# Rename Columns to be able to merge
school_df = school_data.rename(columns={"name":"school_name"})
student_df = student_data.rename(columns={"school":"school_name"})

student_df.head()

# Combine the data into a single dataset
school_data_complete = pd.merge(school_df, student_df, how="left", on=["school_name"])

school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of schools
total_schools = school_df["School ID"].count()

#Calculate the total number of students
total_students = school_df["size"].sum()

# Calculate the total budget
total_budget = school_df["budget"].sum()

# Calculate the average math score 
avgmathscore = school_data_complete["math_score"].mean()

# Calculate the average reading score
avgreadingscore = school_data_complete["reading_score"].mean()

# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overallpassrate = (avgmathscore + avgreadingscore)/2

# Calculate the percentage of students with a passing math score (70 or greater)
passingmath = school_data_complete.query('math_score >70')["School ID"].count()/ total_students*100

# Calculate the percentage of students with a passing reading score (70 or greater)
passingreading = school_data_complete.query('reading_score >70')["School ID"].count()/ total_students*100

# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total Schools":[total_schools],
                                   "Total Students":[total_students],
                                   "Total Budget":[total_budget],
                                   "Average Math Score":[avgmathscore],
                                   "Average Reading Score":[avgreadingscore],
                                   "% Passing Math":[passingmath],
                                   "% Passing Reading":[passingreading],
                                   "% Overall Passing Rate":[overallpassrate]})

district_summary_df = district_summary_df[["Total Schools",
"Total Students",
"Total Budget",
"Average Math Score",
"Average Reading Score",
"% Passing Math",
"% Passing Reading",
"% Overall Passing Rate"]]


district_summary_df
```

## School Summary

```python
# Create an overview table that summarizes key metrics about each school, including:
#   School Name
#   School Type
#   Total Students
#   Total School Budget
#   Per Student Budget
#   Average Math Score
#   Average Reading Score
#   % Passing Math
#   % Passing Reading
#   Overall Passing Rate (Average of the above two)


grouped_by_school  = school_data_complete.groupby(['school_name', "type"])

students_group = grouped_by_school["Student ID"].count()
budget_group = grouped_by_school["budget"].mean()
per_stu_budget =  (budget_group/students_group)
avgmathscore_group = grouped_by_school["math_score"].mean()
avgreadingscore_group = grouped_by_school["reading_score"].mean()
passmath_group = school_data_complete.query('math_score >70')["School ID"].count()/ students_group * 100
passreading_group = school_data_complete.query('reading_score >70')["School ID"].count()/ students_group * 100
overallpassrate_group = (avgmathscore_group + avgreadingscore_group)/2
  
# Create a dataframe to hold the above results
grouped_school_data_df = pd.DataFrame({"Total Students": students_group,
                                       "Total School Budget": budget_group, 
                                       "Per Student Budget":per_stu_budget,
                                       "Average Math Score": avgmathscore_group,
                                       "Average Reading Score": avgreadingscore_group,   
                                       "% Passing Math":passmath_group,
                                       "% Passing Reading":passreading_group,
                                        "Overall Passing Rate":overallpassrate_group})

grouped_school_data_df = grouped_school_data_df[[  "Total Students",
    "Total School Budget",
    "Per Student Budget",
    "Average Math Score",
    "Average Reading Score",
     "% Passing Math",   
     "% Passing Reading",                                             
    "Overall Passing Rate"]]


grouped_school_data_df



grouped_school_data_df.head(10)


```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
by_grade_read_group = school_data_complete.groupby(["school_name","grade"])["reading_score"].mean()
by_grade_read_group_df = pd.DataFrame(by_grade_read_group).reset_index()
by_grade_read_score = by_grade_read_group_df.pivot_table(values="reading_score",index="school_name",columns="grade")
by_grade_read_score = pd.DataFrame(by_grade_read_score)
by_grade_read_score
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
spending_groups = ["<$585", "$585-615", "$615-645", "$645-675"]

# Bin Data
grouped_school_data_df["Per Student Budget"] = pd.cut(per_stu_budget, spending_bins, labels=spending_groups)

# Get scores and percentages based on the Per Student Budget
score_spend_math = grouped_school_data_df.groupby(["Per Student Budget"]).mean()["Average Math Score"]
score_spend_read = grouped_school_data_df.groupby(["Per Student Budget"]).mean()["Average Reading Score"]
score_spend_pass_math = grouped_school_data_df.groupby(["Per Student Budget"]).mean()["% Passing Math"]
score_spend_pass_read = grouped_school_data_df.groupby(["Per Student Budget"]).mean()["% Passing Reading"]
score_spend_overall_pass = (score_spend_pass_math + score_spend_pass_read) / 2

# Create Data Frame to hold information
scores_by_school_spending = pd.DataFrame({"Average Math Score": score_spend_math,
                                                  "Average Reading Score": score_spend_read,
                                                  "% Passing Math": score_spend_pass_math,
                                                  "% Passing Reading": score_spend_pass_read, 
                                                  "Overall Passing Rate": score_spend_overall_pass})

scores_by_school_spending = scores_by_school_spending[["Average Math Score",
                                                                       "Average Reading Score", 
                                                                       "% Passing Math", 
                                                                       "% Passing Reading", 
                                                                       "Overall Passing Rate"]]

scores_by_school_spending.fillna(0)
scores_by_school_spending.head()
```
